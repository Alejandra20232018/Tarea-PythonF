```python
def getDFDataTypes(dataset):
    continuas = []
    discretas = []
    categoricas = []
    for columnName in dataset.columns.values:
        tipoDato = dataset[columnName].dtype
        if((tipoDato == 'int64') or (tipoDato == 'float64')):
            if(len(dataset[columnName].unique()) <= 30):
                discretas.append(columnName)
            else:
                continuas.append(columnName)
        else:
            categoricas.append(columnName)
    return continuas, discretas, categoricas
```

```python
continuas, discretas, categoricas = getDFDataTypes(data)
```

```python
continuas
```

```python
for continueVar in continuas:
     print(continueVar)
     print(f"Media: {data[continueVar].mean()} ")
     print(f"Mediaana: {data[continueVar].median()} ")
     print(f"Maximo: {data[continueVar].max()} ")
     print(f"Minimo: {data[continueVar].min()} ")
     print(f"Desviación: {data[continueVar].std()} ")
     print(f"Varianza: {data[continueVar].var()} ")
     print("--------------------------------")
```
